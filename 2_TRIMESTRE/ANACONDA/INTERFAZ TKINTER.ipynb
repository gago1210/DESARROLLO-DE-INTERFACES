# DESARROLLO DE INTERFACES PYTHON CON TKINTER

```python
### Desarrollar ventanas para las aplicaciones de escritorio Python con Tkiner y pyqt es una forma sencilla de 
### permitir la interacción de los usuarios con nuestras aplicaciones.
```

```python
from tkinter import *

root = Tk()

# bd se refiere al padding
root.config(bd=20)
root.title("Ejemplo tkinter grid")
# tamaño anchoxalto+coordenadaX+coordenadaY
# root.geometry("800x600+250+400")
# 0,0 es para que no se pueda cambiar el tamaño de la ventana
root.resizable(1,1)
root.iconbitmap("seta.ico")

imagen = PhotoImage(file="seta.png").subsample(4)

Label(root, image = imagen).grid(row=0,column=1)
# Texto vacío para crear un espacio entre medias. Si no tiene contenido no existe. No vale con ampliar el nr. de row
Label(root, text="").grid(row=1,column=0)
Label(root, text="Etiqueta 1").grid(row=2,column=0)
Label(root, text="Etiqueta 2").grid(row=3,column=0)
Label(root, text="").grid(row=4,column=0)

eti1 = Entry(root)
eti2 = Entry(root)

eti1.grid(columnspan=4, row=2, column=1)
eti2.grid(columnspan=4, row=3, column=1)

Button(root, text="Aceptar").grid(row=5, column=1)
Button(root, text="Cerrar").grid(row=5, column=2)




root.mainloop()
```

```python
from tkinter import *

# Fuera del interfaz se declaran las funciones

def sumar():
    resultado.set(float(num1.get()) + float(num2.get()))

def restar():
    resultado.set(float(num1.get()) - float(num2.get()))

def multi():
    resultado.set(float(num1.get()) * float(num2.get()))

def borrar():
    num1.set("")
    num2.set("")
    resultado.set("")

# VENTANA DE ALTAS
def altas():
    # TK es la ventana inicial de la aplicación. Toplevel son ventanas que genera el objeto por encima de la ventana inicial
    venAltas = Toplevel()
    imagen = PhotoImage(file="seta.png").subsample(2)
    Label(venAltas, image = imagen).pack()
    Label(venAltas, text="Primer número").pack()

    
    venAltas.mainloop()

# Configuración/Propiedades de la ventana
root = Tk()
root.config(bd=30)
root.title("Calculadora")
root.resizable(1,1)

#Declaración de variables
num1 = StringVar()
num2 = StringVar()
resultado = StringVar()

# Como el posicionamiento es vertical usamos pack. grid lo hemos usado cuando teníamos distribución horizontal


# Componentes interfaz
Label(root, text="Primer número").pack()
# textvariable indica cual es la variable en la que se va a guardar el nr
Entry(root, justify="center", textvariable=num1).pack()

Label(root, text="Segundo número").pack()
Entry(root, justify="center", textvariable=num2).pack()

Label(root, text="Resultado").pack()
Entry(root, justify="center", textvariable=resultado, state="disable").pack()

Label(root, text="").pack() # Espacio de línea vacía

Button(root, text="Sumar", command=sumar).pack(side="left")
Button(root, text="Restar", command=restar).pack(side="left")
Button(root, text="Multiplicar", command=multi).pack(side="left")
Button(root, text="Borrar", command=borrar).pack(side="left")

Button(root, text="Altas", command=altas).pack(side="left")





root.mainloop()
```

## INTERFAZ CON SELECCIÓN OPCIONES CHECKBUTTONS

```python
from tkinter import * 

#hacemos la funcion de seleccionar:
def seleccionar():
    cadena = " "
    if(leche.get()):
        cadena += "Con leche"
    else:
        cadena += "Sin leche"
        
    if(azucar.get()):
        cadena += "y azucar"
    else:
        cadena += "y sin azucar"

#ejecucion de pantalla:
pantalla.config(text = cadena)


#PANTALLA RAIZ
root = Tk()
root.title("CAFETERA")
root.config(bd=15) #bd es que quede margen a los lados 

leche = IntVar() # 1 si, 0 no
azucar = IntVar() #1 si, 0 no

imagen = PhotoImage(file="imagen.gif") #insertar imagen
Label(root, image = image).pack(side="left") #que la imagen se quede a la izda

frame = Frame(root)
frame.pack(side ="left")

Label(frame, text="¿Cómo quiere el café?").pack(anchor="w") #insertamos la frase y la centramos a la derecha 
Checkbutton(frame, text="Con leche",variable=leche, onvalue=1, offvalue=0, command=seleccionar).pack(anchor="w") #cuando pulse onvalue, es 1, cuando no lo pulse offvalue
Checkbutton(frame, text="Con azucar",variable=azucar, onvalue=1, offvalue=0, command=seleccionar).pack(anchor="w")

pantalla = Label(frame) #LABEL DENTRO DEL FRAME
pantalla.pack()

#FINAL BUCLE DEL INTERFAZ
root.mainloop()
```

## INTERFAZ EDITOR DE TEXTO 

```python
#librerias importadas:
from tkinter import * 
from tkinter import filedialog as FileDialog #importamos los cuadros de dialogo para importar un fichero
from io import open #nos gestiona los ficheros (abrir, guardar, buscar, etc)

#funciones:
def nuevo():
    global ruta #se pone global para que la variable este disponible en todo el ambito de la funcion, se le pone la ruta del mensaje 
    mensaje.set("Nuevo fichero")
    ruta = " "
    texto.delete(1.0, "end")
    root.title("Mi editor")
    
def abrir():
    global ruta
    mensaje.set("Abrir fichero")
    ruta = FileDialog.askopentilename(
        initialdir='.',
        filetype=(("Ficheros de texto","*.txt"),),
        title="Abrir un fichero de texto") #hasta aqui hemos generado la ruta, con el nombre de la libreria que importamos 
    
    if ruta != " ": #abrimos un fichero, comprueba que no esta vacio
        fichero= open(ruta,"r") #r es read, lectura
        contenido= fichero.read() #abre un fichero en esa ruta 
        texto.delete(1.0, "end") 
        texto.insert("insert", contenido) #insercion del texto en el contenedor de datos 
        fichero.close() #guardamos los datos
        root,title(ruta + " -Python Editor") #nombre del fichero cuando se abra, muestra la ruta 


def guardar():
    mensaje.set("Guardar fichero") #objeto mensaje con el metodo añadir, añadimos una cosa que tenia
    if ruta != "":#guarda solo en el fichero de la ruta
        contenido = texto.get(1.0, "end -1c")  #recoge todo lo que tenga el texto, 1.0 es el formato de ruta y "" fin de fichero ""end on file")
        fichero = open (ruta,"w+") #aqui se va añadiendo, guardando
        fichero.write(contenido) #aqui meto lo que hemos ido generando en el texto.get
        fichero.close() #cerramos el fichero para que se quede ahi
        mensaje.set("Fichero guardado correctamente")
    else: 
        mensaje.set("Fichero no guardado")
        ruta = " " 
        


def guardarcomo():
    global ruta #como vamos a utilizar una ruta otra vez usamos global ruta
    mensaje.set("Guardar fichero como ")
    fichero = FileDialog.asksavefile(title="Guardar fichero", mode="w" , defaultextension=".txt")
    if fichero is not None: #comprobacion de si hay fichero, si el fichero is not none = true 
        ruta = fichero.name #se cre a partir de ask save ask file, me crea el objeto fichero cuyo atributo es name 
        contenido = texto.get(1.0, "end-1c") #coge el componente de texto hasta el final
        fichero = open(ruta, "w+")
        fichero.write(contenido) #escribe el contenido que ha cargado de leer todo el componente texto
        fichero.close() #cierro el fichero para que se grabe
        mensaje.set("Fichero guardado correctamente")
    else #si no hay fichero:
        mensaje.set("Fichero no guardado poorque no existe")
        ruta = " " 




#PANTALLA/INTERFAZ RAIZ
root = Tk()
root.title("Python Editor")

#vamos a crear una ventana con menu interior con una zona de escritura (poner texto, guardar, etc)
menubar = Menu(root)
filemenu = Menu(menubar, tearoff = 0)#para que esté colapsado
#opciones del menu:
filemenu.add_command(label="Nuevo", command = nuevo)
filemenu.add_command(label= "Abrir", command = abrir)
filemenu.add_command(label="Guardar", command = guardar)
filemenu.add_command(label= "Guardar como", command = guardarcomo)
#metemos un separador de las opciones del menu:
filemenu.add_separator()

filemenu.add_command(label="Salir", command = root.destroy)
#menu en cascada:
menubar.add_cascade(menu = filemenu, label = "Archivo")


#ZONA DE EDICION CENTRAL:
#voy a generar el objeto y guardarlo ahi:
texto = Text(root)
#vamos a meter el componente y añadirlo con pack, fill es el relleno:
texto.pack(fill="both", expand = 1)
#configuramos la fuente con respecto a la interfaz (que no tenga borde/padding) y la fuente que se va a escribir dentro(espacio horizontal padx, espacio vertical pady)
texto.config(bd=0, padx=6, pady=4, font= ("Consolas",12))

#monitor de mensaje o display
mensaje = StringVar()
mensaje.set("Bienvenido a Python Editor, tu editor preferido") #añadir a lo anterior
#pantalla donde se va a presentar el mensaje
monitor = Label(root, textvar= mensaje, justify='left')
monitor.pack(side="left")

#Inclusion del menu desplegable en menu superior



root.config(menu = menubar)

root.mainloop()
```
